```python
import numpy as np
import pandas as pd

import os, sys
from pathlib import Path
import glob

from tqdm import tqdm

CUR_DIR = Path.cwd()
DATA_DIR = CUR_DIR / 'data/preprocessed'
DATA_DIR
```

```python
files = []
for filename in glob.glob(os.path.join(DATA_DIR, '*.csv')):
    cur_file_path = os.path.join(os.getcwd(), filename)
    files.append(cur_file_path)
files
```

```python
len(files)
```

```python
df = pd.read_csv(f'{files[0]}', encoding='utf-8')
df
```

```python
months = range(1, 13)
months = [f'{x}월' for x in months]
months
```

```python
prices = []
new_index = []
for i in df.index:
    year = int(df['구분'][i])
    for month in months:
        price = df.at[i, month]
        new_index.append(f'{year}_{month[:-1]}')
        prices.append(price)
```

```python
len(prices)
```

```python
new_index
```

```python
cut = len(str(DATA_DIR))
crop_name = files[0][cut + 1:-4]
crop_name
```

```python
crop_price = {}
for i, file in enumerate(files):
    crop_name = files[i][cut + 1:-4]
    df = pd.read_csv(f'{file}', encoding='utf-8')
    df.구분 = df.구분.astype('int')
    df = df.loc[df.구분 >= 2001]
    
    prices = []
    new_index = []
    for i in df.index:
        year = int(df['구분'][i])
        for month in months:
            price = df.at[i, month]
            new_index.append(f'{year}_{month[:-1]}')
            prices.append(price)
    
    crop_price[crop_name] = prices
```

```python
crop_price.keys()
```

```python
crop_price['감자']
```

```python
df_total = pd.DataFrame.from_dict(crop_price, orient='columns')
old_columns = list(df_total.columns)
df_total['연월'] = new_index
df_total
```

```python
new_columns = ['연월'] + old_columns
df_total = df_total[new_columns]
df_total
```

```python
save_file_name = DATA_DIR / 'summary'
df_total.to_csv(f'{str(save_file_name)}.csv', encoding='utf-8', index=False)
```
