```python
CUR_DIR = Path.cwd()
DATA_DIR = CUR_DIR / 'data/preprocessed'
DATA_DIR
```

```python
crops = ['감자', '고구마', '고추(화건)', '깻잎', '녹두(국산)', '단감', '당근(무세척)', '대파', '들깨(국산)',
       '딸기', '무', '미나리', '방울토마토', '배(신고)', '배추', '백참깨(국산)', '복숭아(백도)',
       '사과(후지)', '생강(국산)', '수박', '시금치', '애호박', '양배추', '양파', '오이(가시)', '적상추',
       '참외', '콩(흰콩국산)', '토마토', '팥(국산)', '포도(캠벨)', '풋고추']
```

```python
for crop in crops:
    path = f'./results/{crop}/predicts_{crop}.csv'
    df = pd.read_csv(path, encoding='utf-8')
    df['MA_5_true'] = df[f'{crop}_true'].transform(lambda x: x.rolling(5).mean())
    df['MA_5_pred'] = df[f'{crop}_pred'].transform(lambda x: x.rolling(5).mean())
    
    df['보험지급여부_true'] = np.where(df['MA_5_true'] >= df[f'{crop}_true'], 1, 0)
    df['보험지급여부_pred'] = np.where(df['MA_5_pred'] >= df[f'{crop}_pred'], 1, 0)
    
    df['빈도_mean_true'] = df['보험지급여부_true'].expanding().mean()
    df['빈도_mean_pred'] = df['보험지급여부_pred'].expanding().mean()
    
    df['빈도_std_true'] = df['보험지급여부_true'].expanding().std()
    df['빈도_std_pred'] = df['보험지급여부_pred'].expanding().std()
    
    df['심도_true'] = np.where(df['보험지급여부_true'] == 1, df[f'{crop}_true'], 0)
    df['심도_pred'] = np.where(df['보험지급여부_pred'] == 1, df[f'{crop}_pred'], 0)
    
    df['심도_mean_true'] = df['심도_true'].expanding().mean()
    df['심도_mean_pred'] = df['심도_pred'].expanding().mean()
    
    df['보험료_true'] = df['빈도_mean_true'] * df['심도_mean_true']
    df['보험료_pred'] = df['빈도_mean_pred'] * df['심도_mean_pred']
    
    df['누적보험료_true'] = df['보험료_true'].expanding().sum()
    df['누적보험료_pred'] = df['보험료_pred'].expanding().sum()
    
    df.to_csv(f'./results/{crop}/보험료_{crop}.csv', encoding='utf-8', index=False)
    T = df.iloc[-1]['누적보험료_true']
    P = df.iloc[-1]['누적보험료_pred']
    print(f'{crop}: {T:.4f}|{P:.4f}|{T-P:.4f}')
    plt.plot(df['보험료_true'], color='r')
    plt.plot(df['보험료_pred'], color='b')
    plt.title(f'보험료_{crop}')
    plt.show()
```
